```python
# В ЭТОМ ДЗ ВЫ БУДЕТЕ АНАЛИЗИРОВАТЬ ДАННЫЕ ОБ АВИАПРОИСШЕСТВИЯХ С УЧАСТИЕМ МОДЕЛЕЙ ДРОНОВ ИЗ ВАШИХ ИСХОДНЫХ ДАННЫХ В .JSON

# =====================================
# ЗАДАНИЕ 1: Классы - декораторы
# =====================================
# TODO 1-1 - Добавить атрибут incidents типа список
# Возьмите код из предыдущего ДЗ
# Добавьте в класс MultirotorUAV атрибут incidents и внесите туда информацию обо
#всех найденных происшествиях для этой модели
# Не забудьте, что атрибут добавляется при помощи декоратора

import json
import os

class Aircraft:
	def __init__(self, weight):
		self._weight = weight

class UAV:
	def __init__(self):
		self._has_autopilot = True

class MultirotorUAV(Aircraft, UAV):
  def __init__(self, weight, model, brand):
    super().__init__(weight)
    UAV.__init__(self)
    self.__weight = weight
    self.__model = model
    self.__brand = brand
    # добавьте приватный атрибут incidents
    self.__incidents = []
    # напишите код декоратора для атрибута incidents. Не забудьте сначала добавить приватный атрибут в класс
  @property
  def incidents(self):
    return self.__incidents
  
  @incidents.setter
  def incidents(self, incidents):
    self.__incidents = incidents

  @property
  def model(self):
    return self.__model

  @property
  def brand(self):
    return self.__brand

	# напишите публичный метод add_incident, который добавляет инцидент в список инцидентов для данной модели дрона
  def add_incident(self, incident):
    self.__incidents.append(incident)

	# напишите публичный метод save_data, который сохраняет информацию о дроне в файл json
  def save_data(self):
    dictionary ={
      "model" : self.__model,
      "weight": self.__weight,
      "brand":  self.__brand,
      "missions": [],
      "incidents": self.__incidents
    }
    path =  "/content/drive/MyDrive/Colab Notebooks"
    name ="/"+self.__class__.__name__+"_"+self.__brand+" "+self.__model+".json" 
    with open(path+name,"w") as f:
      json.dump(dictionary, f, ensure_ascii=True, indent=2)
    print("file "+name)
    print("saved at "+path)

# ВАШ КОД из предыдущего ДЗ, необходимый для решения этого ДЗ (чтение данных о пилотах, сбор информации о дронах и пр.):
# ее ведь не надо делать...
```

```python
# =====================================
# ЗАДАНИЕ 2: Файлы - CSV
# =====================================
# TODO 2-1 - Загрузите информацию об авиапроисшествиях из файла csv
# Проверьте по моделям (названия моделей возьмите из экземпляров класса MultirotorUAV), какие из них участвовали в авиапроисшествиях

# ВАШ КОД чтения данных из файла:
#-------------------------------------------------------------------------------
def Contains(word, sentence):
  for i in range(0, len(sentence)):
    if(len(sentence)-i-1<len(word)):
      return False
    for j in range(0, len(word)):
      if(word[j] != sentence[i+j]):
        break
      if(j == len(word)-1):
        return True
#-------------------------------------------------------------------------------
def get_brand(drone):
  brand = drone.split()[0].casefold()
  if brand == "dji":
    return "DJI"
  return brand.title()
#-------------------------------------------------------------------------------
def get_model(drone):
  split = drone.split()
  out = ""
  for i in range(1, len(split)-1):
    out = out+split[i]+" "
  return out+split[-1]
#-------------------------------------------------------------------------------
drone_list = ["DJI Mavic 2 Pro", "DJI Mavic 2 Zoom", "DJI Mavic 2 Enterprise Advanced", "AUTEL Evo II Pro", "DJI Mini 2", "Autel Evo Nano", "Autel Evo Lite Plus", "Parrot Anafi", "Dji Inspire 2", "DJI Mavic 3", "DJI Mavic Air2s", "Ryze Tello", "Eachine Trashcan"]
drone_weight_list = [903, 900, 920, 980, 249, 249, 600, 540, 1500, 1000, 570, 130, 110]
models = []

for i in range(0,len(drone_list)):
  models.append(MultirotorUAV(drone_weight_list[i], get_model(drone_list[i]), get_brand(drone_list[i])))
  #print("Model"+str(i)+" = MultirotorUAV(drone_weight_list["+str(i)+"], "+get_model(drone_list[i])+", "+get_brand(drone_list[i])+")")

import csv
fileName = "/content/drive/MyDrive/ColabHW3Files/faa_incidents.csv"
with open(fileName) as f:
  reader = csv.reader(f)
  lines_list = list(reader)

for i in range(0, len(models)):
  for j in range(0, len(lines_list)):
    for k in range (0, len(lines_list[j])):
      if Contains(models[i].model.casefold(), lines_list[j][k].casefold()):
        models[i].add_incident(lines_list[j])
        break

for model in models:
  print(model.brand+" "+model.model)
  if len(model.incidents) == 0:
    print("no incidents")
  for incident in model.incidents:
    print(incident)
```
